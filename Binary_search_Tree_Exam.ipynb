```python
class BinarySearchTree:
    def __init__(self):
        self.root=None
        self.size=0
        
        
    def __len__(self):
        return self.size
    
    def __iter__(self):
        return self.root.__iter__()
    
    def length(self):
        return self.size
    
    
    def put(self,key,value):
        if self.root == None:
            self.root = Node(key,value)
            
        else:
            self._put(self.root,key,value)
            
        self.size +=1
            
    
    def _put(self,start,key,value):
        if key > start.key :
            if start.hasRightChild():
                self._put(start.getRightChild(),key,value)
            else:
                new=Node(key,value)
                start.rightChild=new
                new.parent=start
        
        if key < start.key:
            if start.hasLeftChild():
                self._put(start.getLeftChild(),key,value)
            else:
                new=Node(key,value)
                start.leftChild = new
                new.parent = start
                
        if key == start.key:
            start.payload = value
            
            
            
    
    def get(self,key):
        if self.root == None:
            return None
        
        
        else:
            return self._get(self.root,key)
        
        
    def _get(self,start,key):
        if start.key == key:
            return start.payload
        
        if start.key > key:
            if start.hasLeftChild():
                self._get(start.getLeftChild(),key)
            else:
                return None
            
        if start.key < key:
            if start.hasRightChild():
                self._get(start.getRightChild(),key)
            else:
                return None
            
            
            
    def __contains__(self,key):
        return self._get(self.root,key) != None
    
    
    
    
    
    
    
    
    
    
class Node:
    def __init__(self,key,value,left=None,right=None,parent=None):
        self.key=key
        self.leftChild=left
        self.rightChild=right
        self.parent=parent
        self.payload=value
        
        
    def hasLeftChild(self):
        return self.leftChild !=None
    
    def hasRightChild(self):
        return self.rightChild != None
    
    def isRoot(self):
        return self.parent == None
    
    def isLeaf(self):
        return not self.hasLeftChild() and not self.hasRightChild()
    
    
    def isLeftChild(self):
        return self.parent.leftChild == self
    
    def isRightChild(self):
        return self.parent.rightChild == self
    
    def getLeftChild(self):
        return self.leftChild
    
    def getRightChild(self):
        return self.rightChild
    
    def hasAnyChildren(self):
        return self.hasLeftChild() or self.hasRightChild()
    
    def hasBothChildren(self):
        return self.hasLeftChild() and self.hasRightChild()
    
    def replaceNode(self,key,value,left=None,right=None,parent=None):
        self.key=key
        self.leftChild=left
        self.rightChild=right
        self.parent=parent
        self.payload=value
        
        if self.hasLeftChild():
            self.leftChild.parent=self
        if self.hasRightChild():
            self.rightChild.parent=self
            
    
    
```
